```python
class OcularDataset(Dataset):
  def __init__(self, **kwargs):
    self.data = kwargs['data']
    self.labels = kwargs['labels']

  def __len__(self):
    return self.data.shape[0]

  def __getitem__(self, index):
    _x = self.data[index]
    _y = self.labels[index]
    return _x, _y

def get_dataset(root):
  X = []
  y = pd.read_excel('data/labels.xlsx')
  images = []

  for i in range(y.shape[0]):
    filename = "data/images/" + str(y.values[i][0]) + ".jpg"
    im=np.array(Image.open(filename).convert('L')).flatten() / 255.0
    X.append(im)
    images.append(im)

  y = np.array(y)
  y = y[:, 1]
  
  traindata = []
  trainlab =[]
  #oversampling the data with ones
  for i in range(y.shape[0]):
    image = images[i]
    traindata.append(image)
    trainlab.append(y[i])
    if(y[i] == 1):
      traindata.append(image)
      trainlab.append(1)

  X = traindata
  y = trainlab

  # Shuffling the data  
  idx = np.random.permutation(len(X))
  y = np.array(y)[idx.astype(int)]
  X = np.array(X)[idx.astype(int)]

  print(len(X))

  train = np.array(X[:int(len(X)*(5/7))])
  val = np.array(X[int(len(X)*(5/7)):int(len(X)*(6/7))])
  test = np.array(X[int(len(X)*6/7):])

  kwargs = {'data': train, 'labels': y[:int(len(y)*(5/7))]}
  train = OcularDataset(**kwargs)
  
  kwargs = {'data': val, 'labels': y[int(len(y)*(5/7)):int(len(y)*(6/7))]}
  val = OcularDataset(**kwargs)
  
  kwargs = {'data': test, 'labels': y[int(len(y)*(6/7)):]}
  test = OcularDataset(**kwargs)

  return train, val, test

def calc_positives(labels):
  count = 0
  for i in labels:
    count += i
  return count
```

#### Neural Network [4 pts]

Now, implement your three hidden layer neural network. FNet class will represent your neural network. The layer descriptions are as follows:<ul>
    <i>> Input layer will have ReLU activation. You should decide the number of input neurons.</i><br>
    <i>> First hidden layer will have 1024 neuros with ReLU activation </i><br>
    <i>> Second hidden layer will have 256 neuros with ReLU activation </i><br>
    <i> You should decide the number of output neurons and pick a proper activation function for the output layer. </i><br>
</ul>

```python
class FNet(nn.Module):
    '''Define your neural network'''
    def __init__(self, **kwargs): 
      super(FNet, self).__init__()
      
      self.inputLayer = nn.Sequential(
          nn.Linear(32768, 1024),
          nn.ReLU())
      self.hidden1 = nn.Sequential(
          nn.Linear(1024, 256),
          nn.ReLU())
      self.hidden2 = nn.Sequential(
          nn.Linear(256, 16),
          nn.ReLU())
      self.hidden3 = nn.Sequential(
          nn.Linear(16, 2))      
     
    def forward(self, X): 
      layer1 = self.inputLayer(X)
      layer2 = self.hidden1(layer1)
      layer3 = self.hidden2(layer2)
      out    = self.hidden3(layer3)
      return out
```

```python
# got the idea from https://github.com/pytorch/examples/blob/master/imagenet/main.py#L199
class AverageMeter(object):
  def __init__(self):
    self.reset()

  def reset(self):
    self.val = 0
    self.avg = 0
    self.sum = 0
    self.count = 0

  def update(self, val, step=1):
    self.val = val
    self.sum += val * step
    self.count += step
    self.avg = self.sum / self.count
```

```python
max_epoch = 50
train_batch_size = 64
val_batch_size = 32
test_batch_size = 32
# 0.05 - 50 -> 73, train_batch = 64, val_batch = 32
learning_rate = 0.005

use_gpu = torch.cuda.is_available()

train_loader = DataLoader(train_data, batch_size=train_batch_size, shuffle=True)
val_loader   = DataLoader(val_data,  batch_size=val_batch_size,  shuffle=False)

def main(train_results, val_results): # you are free to change parameters
  save_path = 'nn_best.pth'
  
  model = FNet()
  if use_gpu:
    model.cuda()
  criterion = nn.CrossEntropyLoss()

  optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=
                                                0.9, weight_decay=5e-4)
  optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9, weight_decay=5e-04)
  least_loss = 1e9
  
  for epoch in range(max_epoch):
    res = train(epoch, model, criterion, optimizer, train_loader)
    train_res.append(res)
    res = test(model, val_loader, criterion)
    val_res.append(res)
    accuracy, loss = res[0], res[1]

    if loss < least_loss:
      torch.save(model.state_dict(), save_path)
      least_loss = loss

  print(f'Least loss is: {least_loss}')
    
''' Train your network for a one epoch '''
def train(epoch, model, criterion, optimizer, loader): # you are free to change parameters
  batch_time = AverageMeter()
  accuracies = AverageMeter()
  losses = AverageMeter()
  vakit = time.time()

  model.train()
  for i, (data, labels) in enumerate(loader):
    if use_gpu:
      data, labels = data.cuda(), labels.cuda()

    optimizer.zero_grad()
    output = model(data.float())   
    loss = criterion(output, labels)
    loss.backward()
    optimizer.step()

    _, predicted = torch.max(output, 1)

    accuracy = sum([1 if item[0] == item[1] else 0 for item in zip(predicted, labels)]) / len(data)

    accuracies.update(accuracy, data.size(0))
    losses.update(loss.item(), data.size(0))

    batch_time.update(time.time() - vakit)
    vakit = time.time()

    print(f'Epoch: [{epoch}]\tBatch: [{i}]\t Loss: {losses.avg*100: .3f}%\t Batch time: {batch_time.avg: .3f}s\t Accuracy: {accuracies.avg*100: .3f}%')

  return accuracies.avg, losses.avg

''' Test&Validate your network '''
def test(model, loader, criterion): # you are free to change parameters
  model.eval()

  batch_time = AverageMeter()
  accuracies = AverageMeter()
  losses = AverageMeter()

  with torch.no_grad():
    vakit = time.time()

    for i, (data, labels) in enumerate(loader):
      if use_gpu:
        data, labels = data.cuda(), labels.cuda()

      output = model(data.float())
      loss = criterion(output, labels)
      _, predicted = torch.max(output, 1)
      accuracy = sum([1 if item[0] == item[1] else 0 for item in zip(predicted, labels)]) / len(data)
      accuracies.update(accuracy, data.size(0))
      losses.update(loss.item(), data.size(0))
      batch_time.update(time.time() - vakit)
      vakit = time.time()

    print(f'\tVal_time: {batch_time.avg: .3f}\t Accuracy: {accuracies.avg * 100 : .3f}%\t')

    return accuracies.avg, losses.avg

```

#### Convolutional Neural Network [8 pts]

Now implement your CNN. ConvNet class will represent your convolutional neural network. Implement 3 layers of convolution: 
1. <i>> 8 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br> 
    <i>> ReLU </i><br>
2. <i>> 16 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>   
3. <i>> 32 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>

As a classification layer, you need to add only one more fully-connected layer at the end of the network. You need to choose the appropriate input and output neuron sizes and the activation function.

```python
class ConvNet(nn.Module):
    '''Define your neural network'''
    def __init__(self, **kwargs): # you can add any additional parameters you want 
    # You should create your neural network here
        super(ConvNet, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(3, 8, kernel_size=3, stride=1, padding=0),
            nn.ReLU())
        self.layer2 = nn.Sequential(
            nn.Conv2d(8, 16, kernel_size=3, stride=1, padding=0),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)))
        self.layer3 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=0),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)))
        self.out = nn.Linear(32*30*62, 2)
     
    def forward(self, input): # you can add any additional parameters you want
    # Forward propagation implementation should be here
        in_layer = self.layer1(input)
        layer1 = self.layer2(in_layer)
        layer2 = self.layer3(layer1)        
        layer2 = layer2.reshape(-1,32*30*62)
        out = self.out(layer2)
        return out

```

```python
# HINT: note that your training time should not take many days.

#TODO:
#Pick your hyper parameters
max_epoch = 80
train_batch_size = 64
val_batch_size = 16
test_batch_size = 16
learning_rate = 0.003

use_gpu = torch.cuda.is_available()

def main(train_res, val_res):
    train_loader = DataLoader(train_data, batch_size=train_batch_size, shuffle=True)
    val_loader = DataLoader(val_data, batch_size=val_batch_size)
    test_loader = DataLoader(test_data, batch_size=test_batch_size)
    
    save_path = "cnn_best.pth"

    model = ConvNet()
    criterion = nn.CrossEntropyLoss()

    if use_gpu:
      model.cuda()

    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9, weight_decay=5e-04)
    least_loss = 1e9
    
    for epoch in range(max_epoch):
      res = train(epoch, model, criterion, optimizer, train_loader)

      train_res.append(res)
      res = test(model, val_loader, criterion)
      val_res.append(res)

      loss = res[1]

      if loss < least_loss:
        torch.save(model.state_dict(), save_path)
        least_loss = loss

    print(f'Least loss is: {least_loss}')

def train(epoch, model, criterion, optimizer, loader): 
    model.train()
    
    batch_time = AverageMeter()
    accuracies = AverageMeter()
    losses = AverageMeter()

    vakit = time.time()

    for i, (data, labels) in enumerate(loader):
      if use_gpu:
        data, labels = data.cuda(), labels.cuda()
      
      data = data.permute(0,3,1,2)

      output = model(data.float())
      loss = criterion(output, labels)
      _, predicted = torch.max(output, 1)
      acc = sum([1 if item[0] == item[1] else 0 for item in zip(predicted, labels)]) / len(data)
      accuracies.update(acc, data.size(0))
      losses.update(loss.item(), data.size(0))
        
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
        
      batch_time.update(time.time() - vakit)
      vakit = time.time()

      print(f'Epoch: [{epoch}][{i}]\t Loss: {losses.avg*100: .3f}%\t Batch time: {batch_time.avg: .3f}s\t Accuracy: {accuracies.avg*100: .3f}%')
    
    return accuracies.avg, losses.avg


''' Test&Validate your network '''
def test(model, loader, criterion): # you are free to change parameters

    model.eval()
    batch_time = AverageMeter()
    accuracies = AverageMeter()
    losses = AverageMeter()

    with torch.no_grad():
        vakit = time.time()
        for batch_idx, (data, labels) in enumerate(loader):
          if use_gpu:
            data, labels = data.cuda(), labels.cuda()
          
          data = data.permute(0,3,1,2)

          output = model(data.float())
          labels = labels.long()
          loss = criterion(output, labels)
          _, predicted = torch.max(output, 1)
          acc = sum([1 if item[0] == item[1] else 0 for item in zip(predicted, labels)]) / len(data)
          accuracies.update(acc, data.size(0))
          losses.update(loss.item(), data.size(0))
          
          batch_time.update(time.time() - vakit)
          vakit = time.time()
          
        print(f'\tVal_time: {batch_time.avg: .3f}\t Accuracy: {accuracies.avg * 100 : .3f}%\t')
        
        return accuracies.avg, losses.avg
```
